```python
from transformers import Wav2Vec2Processor, Wav2Vec2ForCTC
import torch

from torchinfo import summary
```

```python
model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-large-960h-lv60-self")
processor = Wav2Vec2Processor.from_pretrained("facebook/wav2vec2-large-960h-lv60-self")
```

```python
processor
```

```python
model.to('cuda:1')
```

```python
model.config
```

```python
summary(model, input_size=(12, 4800))
```

```python
a = torch.rand([2, 80000])
```

```python
a = a.to('cuda')
```

```python
input = processor(a, return_tensors="pt", padding="longest").input_values
```

```python
input.shape
```

```python
out = model.wav2vec2(input[0].to('cuda'))
```

```python
out.last_hidden_state.shape
```
